```python
#Change the variable degree of Gaussian Blur, and change the mode (train or val)
sigma= 7
mode= "val"

for root, dirs, files in os.walk("/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/clean/"+ mode):
    progress=0
    #loop through each of the 21 categories
    for name in dirs:
      num= 0
      progress+=1
      print("\n"+ name+ "\n")
      path= "/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/clean/"+ mode+ "/"+name
      for root, dirs, files in os.walk(path):
        #loop through each of the images in each category
        for imgpath in files:
          print(imgpath+ " "+ str(progress)+"/21")
          path2= os.path.join(root,imgpath)
          img= cv2.imread(path2)
          #conditions to match numbering conventions of original dataset
          #print(img.shape)
          if(mode== "train"):
            if(num<10):
              smoothed_image= cv2.GaussianBlur(img, (0,0), sigma)
              #smoothed_image= img_as_ubyte(gaussian(img, sigma, mode= 'constant', cval=0.0))
              cv2.imwrite("/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/CV2GBlur"+ str(sigma)+ "/"+ mode+ "/"+ name + "/" + name + str(0) + str(num)+".tif", smoothed_image)
              print("a")
            elif(num<42):
              smoothed_image= cv2.GaussianBlur(img, (0,0), sigma)
              #smoothed_image= img_as_ubyte(gaussian(img, sigma, mode= 'constant', cval=0.0))
              cv2.imwrite("/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/CV2GBlur"+ str(sigma)+ "/"+ mode+ "/"+ name + "/" + name + str(num)+".tif", smoothed_image)
              print("b")
            else:
              num2= num + 24
              smoothed_image= cv2.GaussianBlur(img, (0,0), sigma)
              #smoothed_image= img_as_ubyte(gaussian(img, sigma, mode= 'constant', cval=0.0))
              cv2.imwrite("/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/CV2GBlur"+str(sigma)+ "/"+ mode+ "/"+ name + "/" + name + str(num2)+".tif", smoothed_image)
              print("c")
            num+=1
          elif(mode=="val"):
            num2= num+42
            smoothed_image= cv2.GaussianBlur(img, (0,0), sigma)
            #smoothed_image= img_as_ubyte(gaussian(img, sigma, mode= 'constant', cval=0.0))
            cv2.imwrite("/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/CV2GBlur"+str(sigma)+ "/"+ mode+ "/"+ name + "/" + name + str(num2)+".tif", smoothed_image)
            print("d")
            num+=1
          
      
```

Another method- with glob

```python
images_list= []
SIZE= 512
path= "/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/clean/train/airplane/*"

for file in glob.glob(path):
  print(file)
  img= cv2.imread(file)
  img= cv2.resize(img, (SIZE, SIZE))
  images_list.append(img)

images_list= np.array(images_list)

img_number= 0
for image in range(images_list.shape[0]):
  input_img= images_list[image, :, :]
  smoothed_image= img_as_ubyte(gaussian(input_img, sigma=2, mode= 'constant', cval=0.0))
  cv2.imwrite("/content/gdrive/MyDrive/CyberAI/US_UCMerced_China_images/GBlur2/train/airplane/airplane"+ str(img_number)+".tif", smoothed_image)
 
```
